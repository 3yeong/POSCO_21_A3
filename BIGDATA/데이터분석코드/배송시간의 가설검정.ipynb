```python
import pandas as pd
```

```python
df = pd.read_csv('./Sales_data_RMF.csv')
```

```python
df
```

```python
df_4 = df[df['배송시간'] <= '4 days'] 
df_5 = df[df['배송시간']>= '5 days']
```

```python
dup_cnt = df_4.groupby('고객번호')['구매일'].nunique()

# 새로운 데이터프레임 생성
new_df_4 = pd.DataFrame({
    '고객번호': dup_cnt.index,
    '구매횟수': dup_cnt.values
})
```

```python
dup_cnt = df_5.groupby('고객번호')['구매일'].nunique()

# 새로운 데이터프레임 생성
new_df_5 = pd.DataFrame({
    '고객번호': dup_cnt.index,
    '구매횟수': dup_cnt.values
})
```

```python
new_df_4
```

```python
new_df_5
```

```python
from scipy import stats

t_test1 = stats.ttest_ind(new_df_4['구매횟수'], new_df_5['구매횟수'])
print(t_test1)
```

- pvalue 값 거의 0이므로 귀무가설 기각
- 따라서 배송시간(1~4, 5이상) 에 따른 재구매횟수의 차이가 있다
