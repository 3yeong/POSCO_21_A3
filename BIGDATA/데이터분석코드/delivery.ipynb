```python
import pandas as pd

df = pd.read_csv('./New_Sales_data.csv')
df
```

```python
df['배송완료일'] = pd.to_datetime(df['배송완료일'])
df['배송시작일'] = pd.to_datetime(df['배송시작일'])
```

```python
df['배송시간'] = (df['배송완료일'] - df['배송시작일']).dt.days
```
