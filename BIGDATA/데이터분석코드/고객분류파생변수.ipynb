```python
import pandas as pd

df = pd.read_csv('./New_Sales_data.csv')
```

```python
df = df.drop('Unnamed: 0', axis = 1)
```

```python
# df.head()
```

```python
df['현재날짜'] = '2022-08-08'
```

```python
df['현재날짜'] = pd.to_datetime(df['현재날짜'])
```

```python
df['배송시작일'] = pd.to_datetime(df['배송시작일'])
df['배송완료일'] = pd.to_datetime(df['배송완료일'])
```

```python
df['구매일'] = pd.to_datetime(df['구매일'])
```

```python
# df.head()
```

```python
df['구매일로부터기간'] = df['현재날짜'] - df['구매일']
# df.head()
```

```python
df = df.drop_duplicates(['고객번호'], keep = 'last')
len(df)
```

```python
member_df = pd.read_csv('./Member_data.csv')
```

```python
need_df = df[['고객번호', '구매일로부터기간']]
need_df
```

```python
member_df = pd.merge(member_df, need_df, on='고객번호', how='left')
```

```python
member_df
```

```python
member_df['고객분류'] = ''
```

```python
def non_buy(x) :
    return x.replace('NaT', '999999')
```

```python
# 'days' 문자열 제거 및 정수형 변환 함수 정의
def remove_days(x):
    return int(x.replace(' days', ''))
```

```python
# member_df['구매일로부터기간'] = member_df['구매일로부터기간'].astype(str)
# member_df['구매일로부터기간'] = member_df['구매일로부터기간'].apply(remove_days)
```

```python
member_df['구매일로부터기간'] = member_df['구매일로부터기간'].astype(str)
member_df['구매일로부터기간'] = member_df['구매일로부터기간'].apply(non_buy)
member_df['구매일로부터기간'] = member_df['구매일로부터기간'].apply(remove_days)
```

```python
member_df
```

```python
member_df[member_df['구매일로부터기간'] == 999999]
```

```python
# 구매 기간에 따라 고객 분류 함수 정의
def classify_customer(purchase_period):
    if purchase_period <= 90 :
        return '현재고객'
    elif purchase_period <= 180 :
        return '재구매가능성 고객'
    elif purchase_period <= 365 :
        return '잠재이탈고객'
    elif purchase_period <= 600 :
        return '이탈고객'
    else : return '미구매고객'
```

```python
# apply 함수를 사용하여 '고객분류' 열 생성
member_df['고객분류'] = member_df['구매일로부터기간'].apply(classify_customer)
```
